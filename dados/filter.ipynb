```python
import pandas as pd
from unidecode import unidecode
```

```python
df = pd.read_csv('./output.csv',  on_bad_lines='skip')
df.columns = ['frequency', 'words']
df = df[df['words'].str.len() == 5]
df.head(20)
```

```python
df['words'] = df['words'].apply(unidecode)
```

```python
df['words'] = df['words'].str.lower()
df = df.drop_duplicates(subset='words')
df.head()
```

```python
df.to_csv('./palavras_filtradas.csv', index=False)
```
